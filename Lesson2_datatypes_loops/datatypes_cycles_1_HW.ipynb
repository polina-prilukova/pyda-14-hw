## Задание 1
Дана переменная, в которой хранится слово из латинских букв. Напишите код, который выводит на экран:
- среднюю букву, если число букв в слове нечетное;
- две средних буквы, если число букв четное.

Примеры работы программы:  
1. `word = 'test'`

Результат:  
`es`

2.     
`word = 'testing'`

Результат:  
`t`

````python
word = 'polina'
````

````python
middle_index = len(word) // 2 
# для слов нечетной длины достаточно взять букву по полученному индексу, 
# для четной длины - срез +- 1 от индекса 
# (индекс-1 - это левая средняя буква, сам индекс - правая средняя, индекс+1 - граница среза, которая не входит в результат)
if len(word) % 2 == 0:
    print(word[middle_index - 1 : middle_index + 1])
else:
    print(word[middle_index])
    
````

## Задание 2
Напишите программу, которая последовательно запрашивает у пользователя числа (по одному за раз) и после первого нуля выводит сумму всех ранее введенных чисел.

Примеры работы программы:  
1.   

```
Введите число:  
1

Введите число:  
4

Введите число:  
6

Введите число:  
0
```

Результат:  
`11`

2.      

```
Введите число:  
0
```  

Результат:  
`0`

````python
number = int(input('Введите число: '))
summ = number
while number != 0:
    number = int(input('Введите число: '))
    summ += number
print(summ)
    
````

Результат:  
`Внимание, кто-то может остаться без пары!`

````python
boys = ['Peter', 'Alex', 'John', 'Arthur', 'Richard', 'Romeo']
girls = ['Kate', 'Liza', 'Kira', 'Emma', 'Trisha', 'Juliet']
````

````python
if len(boys) != len(girls):
    print('Внимание, кто-то может остаться без пары!')
else:
    boys.sort()
    girls.sort()
    print('Идеальные пары: ')
    for boy, girl in zip(boys, girls):
        print(f'{boy} и {girl}')
````

## Задание 4

У нас есть список, содержащий информацию о среднедневной температуре в Фаренгейтах за произвольный период по странам (структура данных в примере).
Необходимо написать код, который рассчитает среднюю температуру за период в Цельсиях(!) для каждой страны.

Пример работы программы:  
```
countries_temperature = [
    ['Thailand', [75.2, 77, 78.8, 73.4, 68, 75.2, 77]],
    ['Germany', [57.2, 55.4, 59, 59, 53.6]],
    ['Russia', [35.6, 37.4, 39.2, 41, 42.8, 39.2, 35.6]],
    ['Poland', [50, 50, 53.6, 57.2, 55.4, 55.4]]
]
```

Результат:  
```
Средняя температура в странах:
Thailand  -  23.9 С
Germany  -  13.8 С
Russia  -  3.7 С
Poland  -  12.0 С
```

````python
# Воспользуемся формулой (Фаренгейт — 32) : 1,8 = Цельсий
````

````python
countries_temperature = [
    ['Thailand', [75.2, 77, 78.8, 73.4, 68, 75.2, 77]],
    ['Germany', [57.2, 55.4, 59, 59, 53.6]],
    ['Russia', [35.6, 37.4, 39.2, 41, 42.8, 39.2, 35.6]],
    ['Poland', [50, 50, 53.6, 57.2, 55.4, 55.4]]
]
````

````python
print('Средняя температура в странах: ')
for country, temp_list in countries_temperature:
    temp_sum = 0
    for temp in temp_list:
        temp = (temp - 32) / 1.8
        temp_sum += temp
    if len(temp_list) != 0:
        print(f'{country} - {round(temp_sum/len(temp_list), 1)} C')
    else:
        print(f'{country} - нет данных')
    
````

## Задание 5 (необязательное)

Дан поток логов по количеству просмотренных страниц для каждого пользователя. Вам необходимо написать алгоритм, который считает среднее значение просмотров на пользователя. 
Т. е. надо посчитать отношение суммы всех просмотров к количеству уникальных пользователей.

Примеры работы программы:  
1.  

```
stream = [
    '2018-01-01,user1,3',
    '2018-01-07,user1,4',
    '2018-03-29,user1,1',
    '2018-04-04,user1,13',
    '2018-01-05,user2,7',
    '2018-06-14,user3,4',
    '2018-07-02,user3,10',
    '2018-03-21,user4,19',
    '2018-03-22,user4,4',
    '2018-04-22,user4,8',
    '2018-05-03,user4,9',
    '2018-05-11,user4,11',
]
```

Результат:  
`Среднее количество просмотров на уникального пользователя: 23.25`
2.     

```
stream = [
    '2018-01-01,user100,150',
    '2018-01-07,user99,205',
    '2018-03-29,user1001,81'
]
```

Результат:  
`Среднее количество просмотров на уникального пользователя: 145.33`

````python
stream = [
    '2018-01-01,user1,3',
    '2018-01-07,user1,4',
    '2018-03-29,user1,1',
    '2018-04-04,user1,13',
    '2018-01-05,user2,7',
    '2018-06-14,user3,4',
    '2018-07-02,user3,10',
    '2018-03-21,user4,19',
    '2018-03-22,user4,4',
    '2018-04-22,user4,8',
    '2018-05-03,user4,9',
    '2018-05-11,user4,11',
]

# stream = [
#     '2018-01-01,user100,150',
#     '2018-01-07,user99,205',
#     '2018-03-29,user1001,81'
# ]
````

````python
sum_views = 0
unique_users = []
for raw in stream:
    log_list = raw.split(',')
    # Будем считать, что данные корректны, и одна строка всегда состоит из 3х элементов
    # тогда достаточно к общей сумме прибавлять элемент по индексу [2], 
    # а в список уникальных пользователей помещать элемент [1], если его там еще нет
    sum_views += int(log_list[2]) 
    if log_list[1] not in unique_users:
        unique_users.append(log_list[1])
    
if len(unique_users) != 0:
    print(f'Среднее количество просмотров на уникального пользователя: {round(sum_views/len(unique_users), 2)}')
else:
    print('Список уникальных пользователей пуст')

    
````

## Задание 6 (необязательное)

Необходимо у пользователя запрашивать набор чисел разделенных пробелом. В результате должны выводиться числа, которые повторяются в вводе более одного раза.

Примеры работы программы:  
1.  

```
Введите числа:
4 8 0 3 4 2 0 3
```

Результат:  
`0 3 4`
2.     

```
Введите числа:
1 1 2 2 3 3
```

Результат:  
`1 2 3`
3.     

```
Введите числа:
1 1 1 1 1 2 2 2
```

Результат:  
`1 2`

````python
num_string = input('Введите числа через пробел: ')
num_list = num_string.split(' ')
result_list = []
for num in num_list:
    #если число вхождений num больше одного и этого элемента еще нет в результирующем списке, он нам подходит
    if (num_list.count(num) > 1) and (num not in result_list):  
        result_list.append(num)

if len(result_list) != 0: 
    print(' '.join(result_list))
else:
    print('Повторяющихся элементов не найдено!')

    
````
