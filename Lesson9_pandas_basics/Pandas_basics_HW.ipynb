**Задание 2**

По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
power_df = pd.read_csv('power.csv')
power_df.head()
```

```python
# отфильтруем датасет в соответствии с условиями задачи
power_filtered = power_df[(power_df['country'].isin(['Lithuania','Latvia','Estonia']))&
        (power_df['category']).isin([4, 12, 21])&
        ((2005 <= power_df['year']) & (power_df['year'] <= 2010))&
        (power_df['quantity'] >= 0)]

power_filtered
```

```python
# посмотрим суммарное количество потребления
power_filtered['quantity'].sum()
```

```python
# посмотрим в разрезе стран и категорий
power_filtered[['country', 'category', 'quantity']].groupby(['country', 'category']).agg(sum)
```
