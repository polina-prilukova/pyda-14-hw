## Задание 2

Дана переменная, в которой хранится четырехзначное число (год). Необходимо написать программу, которая выведет, является ли данный год високосным или обычным.

Пример работы программы:
1.  

```
year = 2020
```
Результат:  
`Високосный год` 

2.  

```
year = 2019
```
Результат:  
`Обычный год` 

````python
# Год считается високосным, если:
#  год, номер которого кратен 400, — високосный;
#  остальные годы, номер которых кратен 100, — невисокосные (например, годы 1700, 1800, 1900, 2100);
#  остальные годы, номер которых кратен 4, — високосные.
````

````python
year = int(input('Введите год: '))
if year < 1000 or year > 9999:
    print('Год некорректный. Необходимо четырехзначное число')
else:
    if year % 400 == 0 or (year % 4 == 0 and year % 100 != 0):
        print('Високосный год')
    else:
        print('Обычный год')
        
````

Дана переменная, в которой хранится шестизначное число (номер проездного билета). Напишите программу, которая будет определять, является ли данный билет "счастливым". Билет считается счастливым, если сумма первых трех цифр совпадает с суммой последних трех цифр номера.   

Примеры работы программы:  
1. ```number = 123456```

Результат:  
`Неасчастливый билет`
2. ```number = 123321```

Результат:  
`Счастливый билет`


````python
number = int(input('введите 6-значное число: '))
if 1000000 <= number < 100000:
    print('это не 6-значное число')
else:
    first_triad = number // 1000
    second_triad = number % 1000
    print(first_triad, second_triad)
    sum1 = first_triad // 100 + first_triad // 10 % 10 + first_triad % 10
    sum2 = second_triad // 100 + second_triad // 10 % 10 + second_triad % 10
    if sum1 == sum2:
        print('Счастливый билет')
    else:
        print('Несчастливый билет')
    
````

## Задание 6 (необязательное)

Напишите программу, которая сможет вычислять площади трех фигур (круг, треугольник и прямоугольник).
Тип фигуры запрашиваем через пользовательский ввод, после чего делаем запрос характеристик фигуры:

- если пользователь выбрал круг, запрашиваем его радиус,
- если треугольник – длины трех его сторон;
- если прямоугольник – длины двух его сторон.

Пример работы программы:
1.  

```
Введите тип фигуры:
Круг

Введите радиус круга:
10
```
Результат:  
`Площадь круга: 314.16`

2.  

```
Введите тип фигуры:
Треугольник

Введите длину стороны A:
2

Введите длину стороны B:
2

Введите длину стороны C:
3
```
Результат:  
`Площадь треугольника: 1.98`

````python
import math
````

````python
figure_type = input('Введите тип фигуры: ').lower()
square = 0
if figure_type == 'круг':
    r = int(input('Введите радиус круга: '))
    square = round(math.pi * (r ** 2), 3)
elif figure_type == 'треугольник':
    a = int(input('Введите длину стороны A: '))
    b = int(input('Введите длину стороны B: '))
    c = int(input('Введите длину стороны C: '))
    p = (a + b + c) / 2
    square = round(math.sqrt(p * (p - a) * (p - b) * (p - c)), 3)
elif figure_type == 'прямоугольник':
    a = int(input('Введите длину стороны A: '))
    b = int(input('Введите длину стороны B: '))
    square = a * b
else:
    print('Тип фигуры не определен')

if square != 0:    
    print(f'Площадь {figure_type}a: {square}')

    
````
